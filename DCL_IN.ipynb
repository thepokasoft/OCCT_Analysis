```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
%matplotlib inline
```

```python
df1 = pd.read_csv('WSOUT/a.csv', skiprows=1, parse_dates=['DateTime']);
df2 = pd.read_csv('WSOUT/b.csv', skiprows=1, parse_dates=['DateTime']);
df3 = pd.read_csv('WSOUT/c.csv', skiprows=1, parse_dates=['DateTime']);
df4 = pd.read_csv('WSOUT/d.csv', skiprows=1, parse_dates=['DateTime']);
df5 = pd.read_csv('WSOUT/e.csv', skiprows=1, parse_dates=['DateTime']);
df6 = pd.read_csv('WSOUT/f.csv', skiprows=1, parse_dates=['DateTime']);
df7 = pd.read_csv('WSOUT/g.csv', skiprows=1, parse_dates=['DateTime']);
df8 = pd.read_csv('WSOUT/h.csv', skiprows=1, parse_dates=['DateTime']);
df9 = pd.read_csv('WSOUT/i.csv', skiprows=1, parse_dates=['DateTime']);
df10 = pd.read_csv('WSOUT/j.csv', skiprows=1, parse_dates=['DateTime']);
df11 = pd.read_csv('WSOUT/k.csv', skiprows=1, parse_dates=['DateTime']);
df12 = pd.read_csv('WSOUT/l.csv', skiprows=1, parse_dates=['DateTime']);
df13 = pd.read_csv('WSOUT/m.csv', skiprows=1, parse_dates=['DateTime']);
df14 = pd.read_csv('WSOUT/n.csv', skiprows=1, parse_dates=['DateTime']);
df15 = pd.read_csv('WSOUT/o.csv', skiprows=1, parse_dates=['DateTime']);
df16 = pd.read_csv('WSOUT/p.csv', skiprows=1, parse_dates=['DateTime']);
df17 = pd.read_csv('WSOUT/q.csv', skiprows=1, parse_dates=['DateTime']);
df18 = pd.read_csv('WSOUT/r.csv', skiprows=1, parse_dates=['DateTime']);
df19 = pd.read_csv('WSOUT/s.csv', skiprows=1, parse_dates=['DateTime']);
df20 = pd.read_csv('WSOUT/t.csv', skiprows=1, parse_dates=['DateTime']);
df21 = pd.read_csv('WSOUT/u.csv', skiprows=1, parse_dates=['DateTime']);

frames = [df1,df2,df3,df4,df5,df6,df7,df8,df9,df10,df11,df12,df13,df14,df15,df16,df17,df18,df19,df20,df21]

df = pd.concat(frames)
print (df.count())
```

```python
fig, axes = plt.subplots(figsize=(16,9), dpi=150)
axes.legend(["label1","label2","label3","label4","Label5"])

df_temp = df


df_monday = df_temp[df_temp['DateTime'].dt.day_name() == 'Monday']
df_monday = df_monday.groupby(['StopName'], as_index=False).count()
df_monday = df_monday[df_monday.DateTime > 25]
# df_monday.sort_values(by=['DateTime'])
axes.scatter(df_monday['DateTime'], df_monday['StopName'])

df_monday = df_temp[df_temp['DateTime'].dt.day_name() == 'Tuesday']
df_monday = df_monday.groupby(['StopName'], as_index=False).count()
df_monday = df_monday[df_monday.DateTime > 25]
# df_monday.sort_values(by=['DateTime'])
axes.scatter(df_monday['DateTime'], df_monday['StopName'])

df_monday = df_temp[df_temp['DateTime'].dt.day_name() == 'Wednesday']
df_monday = df_monday.groupby(['StopName'], as_index=False).count()
df_monday = df_monday[df_monday.DateTime > 25]
# df_monday.sort_values(by=['DateTime'])
axes.scatter(df_monday['DateTime'], df_monday['StopName'])

df_monday = df_temp[df_temp['DateTime'].dt.day_name() == 'Thursday']
df_monday = df_monday.groupby(['StopName'], as_index=False).count()
df_monday = df_monday[df_monday.DateTime > 25]
# df_monday.sort_values(by=['DateTime'])
axes.scatter(df_monday['DateTime'], df_monday['StopName'])

df_monday = df_temp[df_temp['DateTime'].dt.day_name() == 'Friday']
df_monday = df_monday.groupby(['StopName'], as_index=False).count()
df_monday = df_monday[df_monday.DateTime > 25]
# df_monday.sort_values(by=['DateTime'])
axes.scatter(df_monday['DateTime'], df_monday['StopName'])

axes.legend(["Monday","Tuesday","Wednesday","Thursday","Friday"], loc=1)
plt.savefig('DCL_IN.png')
```
